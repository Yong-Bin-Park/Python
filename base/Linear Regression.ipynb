```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#아이리스 데이터 불러오기
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
display(iris.data.shape, iris.target, iris.target_names, iris.feature_names)
```

```python
#산점도 표현
plt.scatter(iris.data[:,0], iris.data[:,1], c=iris.target, alpha=0.3)
plt.colorbar()
```

```python
X = iris.data[:, 2]
y = iris.data[:, 3]
```

```python
#X(입력값)은 항상 2차원 행렬이어야 한다.
X = iris.data[:, [2]] #X = iris.data[:,2].reshape(-1,1)로 해도 됨
```

```python
X.shape, y.shape
```

```python
#선형회귀
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)   #훈련
y_pred = model.predict(X)   #예측
print(y_pred)
```

```python
plt.scatter(X[:,0],y)
```

```python
plt.scatter(X[:,0],y)
plt.plot(X[:,0],y_pred, 'r')
```
